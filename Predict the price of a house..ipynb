### PREDICT THE PRICE OF HOUSE
King County House price dataset needs to be used for this exercise. 
The data represents the details of houses sold in King County, USA in 2014 and 2015. 
From the given features, predict the price of a house.

```python
import numpy as np
import pandas as pd
```

```python
house_data = pd.read_csv("kc_house_data.csv")       #read data from input csv file         
house_data
```

```python
print(house_data.shape)                              #Determine the size of the data
```

```python
                                                     #Determine the columns in the data
print(house_data.columns)
```

```python
                                                    #View information about the data in each columns
print(house_data.info())
```

### Feature Engineering


```python
house_data_df = house_data[ ["price","date", "bedrooms",                                 #selecting features
                                  "bathrooms", "sqft_living", "floors", 
                                  "waterfront", "view", "condition", "grade","zipcode"]]
house_data_df
```

Here price is considered as the target variable and date, bedrooms, bathrooms, sqft_living, floors, waterfront, view, condition, grade, zipcode as input variables or predictors.

```python
house_data_df.loc[:,"year"] = house_data_df["date"].str[0:4]    #extracting year and month from date           
house_data_df.loc[:,"month"] = house_data_df["date"].str[4:6]
house_data_df[["year","month"]]
house_data_df = house_data_df.drop(columns =["date"])
house_data_df
```

```python
house_data_df              #new dataset after splitting of data into month and year
```

```python
cat_features = ["waterfront", "view", "condition", "grade", "year", "month", "zipcode"] #Encoding the categorical features
house_data_df = pd.get_dummies(house_data_df,columns=cat_features)
print(house_data_df.columns)
house_data_df
```

```python
print(house_data.shape)                       #size of the new dataset after categorical values are encoded
```

```python
from sklearn.preprocessing import StandardScaler   #Normalizing the continuous numerical features
scaler = StandardScaler().fit(house_data_df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'floors']]) 
house_data_normalized_values = scaler.transform(house_data_df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'floors']])

#updating the numerical(normalized) data into the dataframe

house_data_normalized_df = pd.DataFrame(house_data_normalized_values, columns = ['price', 'bedrooms', 'bathrooms', 
                                                                                 'sqft_living', 'floors'])
house_data_normalized_df = house_data_normalized_df.join(house_data_df[house_data_df.columns.drop(['price', 'bedrooms', 'bathrooms', 
                                                                                 'sqft_living', 'floors'])])
house_data_normalized_df
```

```python
y = house_data_normalized_df["price"]  #Selecting features and target
x =  house_data_normalized_df[house_data_normalized_df.columns.drop("price")]
x.shape , y.shape   , house_data_normalized_df.shape
```

Splitting databset into train and test data in a ratio of 80:20.

```python
from sklearn.model_selection import train_test_split
xtrain , xtest, ytrain, ytest = train_test_split(x,y,train_size = 0.8, random_state =100)
xtrain.shape , ytrain.shape, xtest.shape, ytest.shape 
```

### Build a model and evaluate its performance using R-squared
The required model is built to predict the house price based on the selected predictors.

```python
from sklearn.linear_model import LinearRegression      
model = LinearRegression()                             #building and training the model
model.fit(xtrain, ytrain)
```

```python
#Evaluating the model on the train and test data for R-Squared score 
train_score=model.score(xtrain,ytrain)
test_score=model.score(xtest,ytest)
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
```

The model's performance is evaluated using the R2 measure. Since both the train and test data provides comparable value for R-Squared measure, we can say that the model is not over-fitting.

 

### Evaluating the model performance using RMSE
Root Mean Squared Error (RMSE) measure can be used to give an estimate of the average error that can be expected from the model.



```python
from sklearn.metrics import mean_squared_error
train_prediction = model.predict(xtrain)         #root mean square error (RMSE) calculation for train data    
RMSE = mean_squared_error(ytrain,train_prediction)**0.5
print("RMSEtrain :", RMSE)
test_prediction = model.predict(xtest)           #root mean square error (RMSE) calculation for test data 
RMSE = mean_squared_error(ytest,test_prediction)**0.5
print("RMSEtest :",RMSE)
```
